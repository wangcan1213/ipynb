```python
import pandas as pd
import numpy as np
```

## Load File

读取数据内容文件，注意没有表头，所以header=None

```python
d = pd.read_csv('Massachusetts_Tracts_Block_Groups_Only/e20195ma0042000.txt', header=None)
```

此时的数据就是纯数据，没有表头

d.head()

读取数据模板文件，并把它的表头赋予刚才的数据内容文件，然后就有表头了。

```python
d_header = pd.read_excel('2019_5yr_Summary_FileTemplates/seq42.xlsx') 
d.columns = d_header.columns
d.head()
```

读取地理文件，类似的，它也没有表头

```python
g = pd.read_csv('Massachusetts_Tracts_Block_Groups_Only/g20195ma.csv', header=None)
g.head()
```

读取地理模板文件，并用它给地理文件赋予表头

```python
g_header = pd.read_excel('2019_5yr_Summary_FileTemplates/2019_SFGeoFileTemplate.xlsx')
g.columns = g_header.columns
g.head()
```

检查数据文件和地理文件的维度，可见行数是不一样的。这里的数据文件只有Tract和Block Group，而地理文件还包括其他层次和系统的空间统计单元

```python
print(d.shape, g.shape)
```

基于LOGRECNO列，合并数据文件和地理文件，合并后只保留我们需要的B15001系列的属性，以及一些跟空间有关的地理信息列

```python
dg = d.merge(g, on='LOGRECNO')
geo_columns = ['STATE', 'COUNTY', 'TRACT', 'BLKGRP','NAME']
use_data_cols = [f'B15001_{str(x+1).zfill(3)}' for x in range(83)] 
dg = dg[geo_columns+use_data_cols]
dg.head()
```

顺便看一下，有哪些地名是数据文件中没有，而地理文件中有的，打印前20个，可以看到第一个就是麻州全州的，第2、3个分别是麻州城镇地区、麻州乡村地区……

```python
name = [x for x in list(g['NAME']) if x not in list(dg['NAME'])]
```

```python
name[:20]
```

把1001、1002、1003这3个Tracts的数据筛出来。打印结果，完全一致

```python
places = [f'Census Tract {x}, Suffolk County, Massachusetts' for x in [1001,1002,1003]]
dg2 = dg.loc[dg['NAME'].isin(places)]
dg2
```
